## Preprocesamiento de base de datos con pandas


```python
from SignalClases import saveData
from SignalClases import saveDataSR

import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
```

```python
# Indices para carpetas a revisar 
numFol = 4
# numFile = [8,4,8,8]
numFile = [3,2,2,3]
numLabel = 8
srate = 10000                                    
```

```python
for a in range(0,numFol):                       # range(0,numFol)
    Foldnum = a
    Filenum = numFile[a]
    
    for lb in range(0,numLabel):            # 0,numLabel
        dataPath = saveData.getData(Foldnum, Filenum, lb)       # carga de dataset inicial 
        df = pd.read_csv(dataPath)
        df = df.drop('Duration',axis=1)
        df = df.drop_duplicates(subset=['indFR F'])
        df = df.drop_duplicates(subset=['indFR I'])
        df = df.reset_index(drop=True)
        df['Duration'] = (df['indFR F']- df['indFR I'])*1/srate
        pathSV = saveDataSR.getData(Foldnum,Filenum,lb)
        df.to_csv(pathSV,sep=',',index=False)  
```

```python
lst = []

for a in range(0,numFol):                       # range(0,numFol)
    for b in range(0,numFile[a]):               # 0,numFile[a]
        for lb in range(7,8):            # 0,numLabel
            Foldnum = a
            Filenum = b
            
            dataPath = saveDataSR.getData(Foldnum, Filenum, lb)       # carga de dataset inicial 
            df = pd.read_csv(dataPath)
            
            mn = df['Duration'].mean()
            lst.append(mn)

med = sum(lst)/ len(lst)
lst_mean.append(med)
```
